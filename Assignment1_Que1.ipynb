```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img=cv2.imread('pikachu.jpg')
img_rgb = cv2.cvtColor(img.astype(np.uint8), cv2.COLOR_BGR2RGB)
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img_gray_resized=cv2.resize(img_gray, (150, 150))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
plt.imshow(img_gray_resized,cmap='gray')
plt.axis('off')
plt.show()
```

```python
#CODE TO MAKE THE IMAGE BLURR

def multiply(img,kernel):    #element wise multiplication applicable only for 3x3 arrays
    arr=np.zeros((3,3))
    for i in range(3):
        for j in range(3):
            arr[i,j]=img[i,j]*kernel[i,j]
    return arr
def blurring_filter(img):
    kernel=np.array([[1,1,1],[1,1,1],[1,1,1]],dtype=np.float32)    #box blur kernel
    kernel=kernel/9
    num_rows,num_col=img.shape
    result_img=np.zeros((num_rows-2,num_col-2))
    for i in range(num_rows-2):
        for j in range(num_col-2):
            sub_img=img[i:i+3,j:j+3]
            arr=multiply(sub_img,kernel)
            result_img[i,j]=arr.sum()
    return result_img

blurred_img=blurring_filter(img_gray_resized)
plt.imshow(blurred_img,cmap='gray')
plt.axis('off')
plt.show()
```

```python
#CODE TO MAKE THE IMAGE SHARP

def multiply(img,kernel):    #element wise multiplication applicable only for 3x3 arrays
    arr=np.zeros((3,3))
    for i in range(3):
        for j in range(3):
            arr[i,j]=img[i,j]*kernel[i,j]
    return arr
def sharpening_filter(img):
    kernel=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]],dtype=np.float32)
    kernel=kernel/kernel.sum()
    num_rows,num_col=img.shape
    result_img=np.zeros((num_rows-2,num_col-2))
    for i in range(num_rows-2):
        for j in range(num_col-2):
            sub_img=img[i:i+3,j:j+3]
            arr=multiply(sub_img,kernel)
            result_img[i,j]=arr.sum()
    return result_img

sharpened_img=sharpening_filter(img_gray_resized)
plt.imshow(sharpened_img,cmap='gray')
plt.axis('off')
plt.show()
```
